Extract page start from: https://kathika.org/stories

```python
# code:
def get_author_and_content(url):
    link = 'https://kathika.org' + url
    response = req.get(link)
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    all_div = soup.find_all('div', class_='story-content ql-editor')
    user_details_div = soup.find_all('div', class_='user-details')
    
    if all_div:
        story_content = all_div[0].text.strip()
    else:
        story_content = None
        
    author_name = None  # Set a default value
    
    for div in user_details_div:
        author_name_element = div.find('h3')
        if author_name_element:
            author_name = author_name_element.get_text(strip=True)
            break  # Break the loop if the author name is found
    
    return author_name, story_content

book_info = {}

# Specify the URL
url = "https://kathika.org/stories"

# Send a GET request to the main URL
response = req.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    all_a_tags = soup.find_all('a')
    
    for a_tag in all_a_tags:
        link = a_tag['href']
        if link[0] != '/':
            continue  # Skip if the link doesn't start with '/'
        
        author_name, story_content = get_author_and_content(link)
        h3_tag = a_tag.find('h3', title=True)
        book_name = h3_tag['title'] if h3_tag else None
        
        if book_name is not None:
            book_info[book_name] = {
                'author': author_name,
                'content': story_content
            }
else:
    print(f"Failed to retrieve content. Status code: {response.status_code}")

print(book_info)
```
